```python
df.rename(columns = {'fixed acidity':'fixed_acidity', 'volatile acidity':'volatile_acidity',
                              'citric acid':'citric_acid',"residual sugar":"residual_sugar","free sulfur dioxide":"free_sulphor_dioxide","total sulfur dioxide":"total_sulfur_dioxide"}, inplace = True)
```

```python
df.columns
```

```python
#now we will handle null values easily.
```

```python
df[df.fixed_acidity.isnull()]
```

```python
df['fixed_acidity'] = df['fixed_acidity'].fillna(df['fixed_acidity'].median())
```

```python
df.fixed_acidity.isnull().sum()
```

```python
#automated the process of filling all the missing values.

for i in df.columns:
    if df[i].isnull().sum() != 0:
        df[i] = df[i].fillna(df[i].median())
        

        
```

```python
df.isnull().sum()
#we have no null values now.
```

```python
## check for duplicated records
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
## segregate numerical and categorical columns

numerical_columns=df.columns[df.dtypes!='object']
categorical_columns=df.columns[df.dtypes=='object']
print("Numerical columns:",numerical_columns)
print('Categorical Columns:',categorical_columns)
```

```python
df[categorical_columns].describe()
```

```python
#giving ranking as per domain knowledge
type_map={"white":"1","red":2}
```

```python
#mapping them to their respected values
df['type']=df['type'].map(type_map)
```
